```python
#reference-since this column doesnot seem to be not very important and it is categorical column.
#so we can either drop the records or impute with the mode.
#now as number of missing is not very less so lets not drop the records.
#lets impute missing values with mode.
```

# imputing for column CNT_FAM_MEMBERS

```python
app_data['CNT_FAM_MEMBERS'].describe()
```

```python
app_data['CNT_FAM_MEMBERS'].quantile([0.9,0.95,0.99])
```

```python
#reference-its a numeric discrete column so obviously we wont use mean to impute.
#we seen 99% value is 5 or under and median is 2.so if we just impute missing values with median,it will be perfect imputation.
#our analysis wont be exaggerated by this imputation.
```

```python
app_data['CNT_FAM_MEMBERS']=app_data['CNT_FAM_MEMBERS'].fillna(app_data['CNT_FAM_MEMBERS'].median())
```

lets verify the change

```python
app_data['CNT_FAM_MEMBERS'].isnull().sum()
#so change applied
```
